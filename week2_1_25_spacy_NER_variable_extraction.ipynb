## Spacy NER for entity extraction on text extracted using pyPDF2 

```python
!pip install pypdf2
from google.colab import drive
drive.mount('/content/drive')

#https://medium.com/analytics-vidhya/how-to-extract-texts-from-pdf-file-and-search-keywords-from-extracted-text-in-python-c5f3d4841f20
from PyPDF2 import PdfFileReader
pdf_path='/content/drive/MyDrive/week1/pdf1.pdf'

pdfFileObj = open(pdf_path, 'rb')
#We opened the example.pdf in binary mode. and saved the file object as pdfFileObj
pdfReader = PdfFileReader(pdfFileObj)
#create an object of PdfFileReader class of PyPDF2 module and  pass the pdf file object & get a pdf reader object.
print(pdfReader.numPages) # will give total number of pages in pdf
```

```python
page_text = []
#ingest data from the pdf 
for i in range(pdfReader.numPages):
#for each page 
  pageObj = pdfReader.getPage(i)
  #text = pageObj.extractText().lower()
  text = pageObj.extractText()
  page_text.append(text)

for i in range(len(page_text)):
  print("Page:" , str(i), ":" , page_text[i], "\n")
```

```python
#convert list to dataframe 
from pandas import DataFrame
df_extracted_pages = DataFrame(page_text, columns =['text'])
```

```python
def listToString(s): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for ele in s: 
      str1 += ele
    
    
    # return string  
    return str1 
```

```python
corpus = listToString(page_text)
```

```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
article = nlp(corpus)
```

##Display default language model entities 

```python
displacy.render(article, style = 'ent', jupyter=True)
```

## Phrase matching to match the phrase "AIRIA HOLDINGS INC." 

```python
import spacy
from spacy.matcher import PhraseMatcher
corpus = listToString(page_text)
nlp = spacy.load("en_core_web_sm")
matcher = PhraseMatcher(nlp.vocab, attr = "LOWER")
terms = ["AIRIA HOLDINGS INC."]
# Only run nlp.make_doc to speed things up
patterns = [nlp.make_doc(text) for text in terms]
matcher.add("TerminologyList", patterns)

doc = nlp(corpus)
matches = matcher(doc)
for match_id, start, end in matches:
    span = doc[start:end]
    print("Starting at:", start, "Ending at:", end, "found:", span.text)
```
